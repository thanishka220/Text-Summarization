```python
#  Install and import only what's needed
!pip install nltk --quiet
import textwrap


import nltk
nltk.download('stopwords', force=True)

from nltk.corpus import stopwords
from collections import defaultdict
import string
import heapq
import re

#  Upload your .txt file
from google.colab import files
uploaded = files.upload()

#  Read the uploaded text file
file_name = list(uploaded.keys())[0]
with open(file_name, 'r', encoding='utf-8') as file:
    text = file.read()

#  Fallback: split sentences using regex
sentences = re.split(r'(?<=[.!?])\s+', text.strip())

#  Fallback: simple regex word tokenizer
def tokenize_words(text):
    return re.findall(r'\b\w+\b', text.lower())

#  Build word frequency dictionary
stop_words = set(stopwords.words('english'))
word_frequencies = defaultdict(int)

for word in tokenize_words(text):
    if word not in stop_words:
        word_frequencies[word] += 1

#  Normalize frequencies
max_freq = max(word_frequencies.values())
for word in word_frequencies:
    word_frequencies[word] /= max_freq

#  Score sentences
sentence_scores = defaultdict(int)
for sentence in sentences:
    for word in tokenize_words(sentence):
        if word in word_frequencies:
            sentence_scores[sentence] += word_frequencies[word]

# Select top 3 important sentences
summary_sentences = heapq.nlargest(3, sentence_scores, key=sentence_scores.get)
summary = ' '.join(summary_sentences)

# Output
# Print original text preview line by line
wrapper = textwrap.TextWrapper(width=80)  # You can adjust width as needed

print("\n Original Text:\n")
for line in text.split('\n'):  # full original text line by line
    wrapped_lines = wrapper.wrap(line)
    for wrapped_line in wrapped_lines:
        print(wrapped_line)

print("\n Summarized Text:\n")
for sent in summary_sentences:
    wrapped_lines = wrapper.wrap(sent)
    for wrapped_line in wrapped_lines:
        print(wrapped_line)
```
